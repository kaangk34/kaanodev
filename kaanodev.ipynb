```python
import math
import random
import matplotlib.pyplot as plt

# Okul numarasının son iki hanesini benim okul numaramın sonu 20 olarak belirledim.
school_number_last_2_digits = 20
top_konum = [0, school_number_last_2_digits]

angle = 30
g = 10  # yer çekimi ivmesi

def initial_speed(lower_speed_limit, upper_speed_limit):
    return (lower_speed_limit + upper_speed_limit) / 2

def horizontal_speed(v0):
    return v0 * math.cos(math.radians(angle))

def vertical_speed(v0):
    return v0 * math.sin(math.radians(angle))

def flight_time(v0y):
    return 2 * v0y / g

def range_x(v0x, t_flight):
    return v0x * t_flight

# Rastgele hedef belirleme
distance = 20_000 + 200 * random.randint(-10, 10)
width_start = distance
width_end = distance + 1_000 + 100 * random.randint(-2, 2)

# Hız aralığı ve atış sayısı
lower_speed_limit = 330
upper_speed_limit = 1800
shots = 0

# Atışların yatay ve dikey bileşenlerindeki değişimleri tutacak listeler
x_list = []
y_list = []

while True:
    shots += 1
    v0 = initial_speed(lower_speed_limit, upper_speed_limit)
    v0x = horizontal_speed(v0)
    v0y = vertical_speed(v0)
    t_flight = flight_time(v0y)
    x_range = range_x(v0x, t_flight)

    if width_start <= x_range <= width_end:
        print(f"Hedefi vurdun!\n{shots}. seferde hedefi vurdun... hedefi vurmak için gerekli hız {v0} m/s'dir")
        break
    elif x_range < width_start:
        print("önüne düştü")
        lower_speed_limit = v0
    else:
        print("uzagına düştü")
        upper_speed_limit = v0

    # Atışın yatay ve dikey bileşenlerindeki değişimleri listelere ekleme
    x_list.append(x_range)
    y_list.append(v0y * t_flight - 0.5 * g * t_flight**2)

# Grafik çizdirme
plt.plot(x_list, y_list)
plt.title("Top Mermisinin Yörüngesi")
plt.xlabel("Mesafe (m)")
plt.ylabel("Yükseklik (m)")
plt.annotate(f"Atış Sayısı: {shots}", xy=(0.05, 0.95), xycoords='axes fraction')

plt.show()
```
